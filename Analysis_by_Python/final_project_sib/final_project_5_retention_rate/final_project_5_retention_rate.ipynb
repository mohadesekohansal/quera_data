<font face="vazirmatn" size=3>

| <b style="color:#0099cc">ویژگی</b> | <b style="color:#0099cc">مقدار</b> |
| :---: | :---: |
| اندازه‌ی شکل | ۱۵ در ۸ اینچ |
| بازه اعداد برای تابع نگاشت رنگ | ۰ تا ۵۰ |
| کد فرمت مورد استفاده | <code dir=ltr>.0f</code> |
| تابع نگاشت رنگ (<code>cmap</code>) | <code>BuGn</code> |

</font>

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["InvoiceDate"] = df["InvoiceDate"].apply(lambda x: x.replace(day=1))

first_date_customer = df.groupby("CustomerId")["InvoiceDate"].min().sort_index()
df["first_date"] = df["CustomerId"].apply(lambda x: first_date_customer.loc[x])
df["duration"] = (df["InvoiceDate"].dt.year - df["first_date"].dt.year) * 12 + (df["InvoiceDate"].dt.month - df["first_date"].dt.month)
df["first_date"] = df["first_date"].dt.date
df
```

```python
plot_data = df.groupby(["first_date","duration"])["CustomerId"].unique().apply(lambda x: len(x)).unstack()
plot_data = round(((plot_data.T / plot_data[0].values).T)*100)
plot_data
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,8))
sns.heatmap(data = plot_data ,cmap="BuGn" , vmin= 0 , vmax=50 , annot= True , fmt= ".0f" )
ax.set_title("Retention rates")
ax.set_xlabel("CohortIndex")
ax.set_ylabel(None)
ax.set_xticklabels(range(1,14))
plt.show()
```
