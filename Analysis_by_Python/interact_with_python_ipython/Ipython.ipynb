ШӘЩ…ШұЫҢЩҶ ЩҫШ§ЫҢШӘЩҲЩҶ ШӘШ№Ш§Щ…Щ„ЫҢ
</font>
</h1>


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШҜШұ Ш§ЫҢЩҶ ШӘЩ…ШұЫҢЩҶ Щ…ЫҢвҖҢШ®ЩҲШ§ЩҮЫҢЩ… ЫҢЪ© ШіШұЫҢ Ш№Щ…Щ„ЫҢШ§ШӘ ШіШ§ШҜЩҮ Ш§ЩҶШ¬Ш§Щ… ШҜЩҮЫҢЩ… ШӘШ§ ШЁШ§ ЩҶШӘвҖҢШЁЩҲЪ© ЩҫШ§ЫҢШӘЩҲЩҶ ЩҲ ЩҲЫҢЪҳЪҜЫҢвҖҢЩҮШ§ЫҢ ШўЩҶ ШЁЩҮШӘШұ ШўШҙЩҶШ§ ШҙЩҲЫҢШҜ.
<br>
ШҜШұ ЩҫШ§ЫҢШӘЩҲЩҶ ШӘШ§ШЁШ№ЫҢ ШЁЩҮ ЩҶШ§Щ… <code>any</code> ЩҲШ¬ЩҲШҜ ШҜШ§ШұШҜ Ъ©ЩҮ ЫҢЪ© <code>Iterator</code> (Щ…Ш«Щ„ ШҜЫҢЪ©ШҙЩҶШұЫҢШҢ Щ„ЫҢШіШӘ ЫҢШ§ ШӘШ§ЩҫЩ„) ШұШ§ Щ…ЫҢвҖҢЪҜЫҢШұШҜ ЩҲ Ш§ЪҜШұ ШӯШҜШ§ЩӮЩ„ ЫҢЪ© Щ…ЩӮШҜШ§Шұ <code>True</code> ШҜШ§ШҙШӘЩҮ ШЁШ§ШҙШҜ ЩҫШ§ШіШ® <code>True</code> ШӘЩҲЩ„ЫҢШҜ Щ…ЫҢвҖҢЪ©ЩҶШҜ (ЫҢШ№ЩҶЫҢ ШӘЩҶЩҮШ§ ЩҲЩӮШӘЫҢ <code>False</code> ШӘЩҲЩ„ЫҢШҜ Щ…ЫҢвҖҢЪ©ЩҶШҜ Ъ©ЩҮ ШӘЩ…Ш§Щ… Щ…ЩӮШ§ШҜЫҢШұ ШўЩҶ <code>False</code> ШЁШ§ШҙЩҶШҜ).
<br>
Щ…Ш«Ш§Щ„вҖҢЩҮШ§ЫҢЫҢ Ш§ШІ Ш§ЫҢЩҶ ШӘШ§ШЁШ№ ШұШ§ ШҜШұ ШіЩ„ЩҲЩ„ ШІЫҢШұ Щ…ЫҢвҖҢШЁЫҢЩҶЫҢШҜ:

</font>
</p>

```python
print(any([True, False, True]))
print(any([False, False, False]))
print(any([0, False]))
print(any([1, 2, 3]))
print(any(["salaam"]))
```


<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ЩӮШіЩ…ШӘ Ш§ЩҲЩ„
</font>
</h2>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШҜШұ ШіЩ„ЩҲЩ„ ШІЫҢШұ ШӘШ§ШЁШ№ЫҢ ШЁЩҶЩҲЫҢШіШҜ Ъ©ЩҮ Ъ©Ш§Шұ ШӘШ§ШЁШ№ <code>any</code> ШұШ§ Ш§ЩҶШ¬Ш§Щ… ШҜЩҮШҜ. ШӘШ§ШЁШ№ ШҙЩ…Ш§ ШЁШ§ЫҢШҜ ЫҢЪ© Щ„ЫҢШіШӘ ШұШ§ ЩҲШұЩҲШҜЫҢ ШЁЪҜЫҢШұШҜ (<code>x</code>) ЩҲ Ш§ЪҜШұ ШҜШұ Щ„ЫҢШіШӘ ШӯШҜШ§ЩӮЩ„ ЫҢЪ© Щ…ЩӮШҜШ§Шұ <code>True</code> ЩҲШ¬ЩҲШҜ ШҜШ§ШҙШӘЩҮ ШЁШ§ШҙШҜ ЩҫШ§ШіШ® <code>True</code> ШұШ§ ШЁШұЪҜШұШҜШ§ЩҶШҜ ЩҲ ШҜШұ ШәЫҢШұ Ш§ЫҢЩҶвҖҢШөЩҲШұШӘ (Ш§ЪҜШұ ШӘЩ…Ш§Щ… Щ…ЩӮШ§ШҜЫҢШұ <code>False</code> ШЁЩҲШҜЩҶШҜ) <code>False</code> ШЁШұЪҜШұШҜШ§ЩҶШҜ. ШҜЩӮШӘ Ъ©ЩҶЫҢШҜ Ъ©ЩҮ ШӘШ§ШЁШ№ ШЁШ§ЫҢШҜ Ш®ШұЩҲШ¬ЫҢ ШұШ§ <code>return</code> Ъ©ЩҶШҜ (ЪҶШ§Щҫ ЩҶЪ©ЩҶШҜ).
</font>
</p>
<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
<span style="color:purple"><b>ЩҶЪ©ШӘЩҮ:</b></span>
ЩӮШ§Ш№ШҜШӘШ§ЩӢвҖҢ ШҜШұЩҲЩҶ Ш§ЫҢЩҶ ШӘШ§ШЁШ№ ЩҶШЁШ§ЫҢШҜ Ш§ШІ Ш®ЩҲШҜ ШӘШ§ШЁШ№ <code>any</code> Ш§ШіШӘЩҒШ§ШҜЩҮ Ъ©ЩҶЫҢШҜШҢ ШЁЩ„Ъ©ЩҮ ШЁШ§ЫҢШҜ ШЁЩҮвҖҢШөЩҲШұШӘ ШҜШіШӘЫҢ ШұЩҒШӘШ§Шұ ШўЩҶ ШұШ§ Ъ©ШҜЩҶЩҲЫҢШіЫҢ Ъ©ЩҶЫҢШҜ.
</font>
</p>


```python
def my_any(x):
    for y in x:
        if y != False:
            return True
    return False
```


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШҜШіШӘЩҲШұШ§ШӘ Ш¬Ш§ШҜЩҲЫҢЫҢ ШҜШ§ШұШ§ЫҢ ШўШұЪҜЩҲЩ…Ш§ЩҶвҖҢЩҮШ§ЫҢ ШЁШіЫҢШ§Шұ Ъ©Ш§ШұШЁШұШҜЫҢвҖҢвҖҢШ§ЫҢ Щ…ЫҢвҖҢШЁШ§ШҙЩҶШҜ Ъ©ЩҮ ШЁШұШ§ЫҢ Щ…ШҙШ§ЩҮШҜЩҮвҖҢЫҢ ШўЩҶвҖҢЩҮШ§ШҢ Щ…ЫҢвҖҢШӘЩҲШ§ЩҶ Ш§ШІ ЩҲЫҢЪҳЪҜЫҢ <code>introspection</code> Ш§ШіШӘЩҒШ§ШҜЩҮ Ъ©ШұШҜ:

</font>
</p>



```python
%timeit?
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ЩҮЩ…Ш§ЩҶвҖҢШ·ЩҲШұ Ъ©ЩҮ ШҜШұ ШҜШұШіЩҶШ§Щ…ЩҮвҖҢЫҢ ЩҫЫҢШҙЫҢЩҶ ЩҮЩ… Щ…Ш·Ш§Щ„Ш№ЩҮ Ъ©ШұШҜЫҢШҜ Ш№Щ„Ш§Щ…ШӘ <code>%</code> ШҜШұ Ш§ШЁШӘШҜШ§ЫҢ Ш§ЫҢЩҶ ШҜШіШӘЩҲШұ ШЁЫҢШ§ЩҶвҖҢЪҜШұ Ш¬Ш§ШҜЩҲЫҢЫҢ (magic) ШЁЩҲШҜЩҶ ШўЩҶ Ш§ШіШӘ ЩҲ Ш№Щ„Ш§Щ…ШӘ <code>?</code> Ш§ЩҶШӘЩҮШ§ ЩҶЫҢШІ ШЁЩҮ Ш§ЫҢЩҶ Щ…Ш№ЩҶЫҢвҖҢШіШӘ Ъ©ЩҮ ЩӮШөШҜ ШҜШ§ШұЫҢЩ… Ш§ШІ Ш§Ш·Щ„Ш§Ш№Ш§ШӘ Ъ©Щ„ЫҢ Ш§ЫҢЩҶ ШҙЫҢШЎ ЫҢШ§ ШӘШ§ШЁШ№ Щ…Ш·Щ„Ш№ ШҙЩҲЫҢЩ….
</font>
</p>


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШҜШұ Ш®ШұЩҲШ¬ЫҢ ШіЩ„ЩҲЩ„ ШЁШ§Щ„Ш§ ШӘЩҲШ¶ЫҢШӯШ§ШӘ ШЁШіЫҢШ§Шұ Щ…ЩҒЫҢШҜ ЩҲ Ъ©Ш§ШұШЁШұШҜЫҢвҖҢШ§ЫҢ ШұШ§ ШҜШұ ШұШ§ШЁШ·ЩҮ ШЁШ§ <code>timeit</code> Щ…ШҙШ§ЩҮШҜЩҮ Щ…ЫҢвҖҢЪ©ЩҶЫҢШҜ.
Щ…Ш«Ш§Щ„вҖҢЩҮШ§ЫҢЫҢ ЩҮЩ… ШЁШұШ§ЫҢ ШўШҙЩҶШ§ЫҢЫҢ ШЁЫҢШҙШӘШұ ШўЩ…ШҜЩҮ Ш§ШіШӘ. Ш§ЪҜШұ ШҜЩӮШӘ Ъ©ЩҶЫҢШҜ Щ…ЫҢвҖҢШЁЫҢЩҶЫҢШҜ Ъ©ЩҮ ШЁШ§ ШӘЩҶШёЫҢЩ… <code dir=ltr>-o</code> Щ…ЫҢвҖҢШӘЩҲШ§ЩҶ Ш®ШұЩҲШ¬ЫҢ ШҜШіШӘЩҲШұ Ш¬Ш§ШҜЩҲЫҢЫҢ <code>timeit</code> ШұШ§ ШҜШұ Щ…ШӘШәЫҢШұЫҢ Ш°Ш®ЫҢШұЩҮ Ъ©ШұШҜ ЩҲ Ш§ШІ ШўЩҶ Ш§Ш·Щ„Ш§Ш№Ш§ШӘ ШЁЫҢШҙШӘШұЫҢ 
Ш§ШіШӘШ®ШұШ§Ш¬ Ъ©ШұШҜ.

</font>
</p>


<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ШіЩҲШ§Щ„ ШҜЩҲЩ…
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШҜШұ Ш§ЫҢЩҶ ЩӮШіЩ…ШӘ ЫҢЪ© Щ„ЫҢШіШӘ ЩҲШұЩҲШҜЫҢ ШҜЩ„Ш®ЩҲШ§ЩҮ Ш§ЩҶШӘШ®Ш§ШЁ Ъ©ШұШҜЩҮ (ШЁЩҮвҖҢШ№ЩҶЩҲШ§ЩҶ Щ…Ш«Ш§Щ„ <code dir=ltr>[1, 2, 3]</code>) ШіЩҫШі ЫҢЪ©вҖҢШЁШ§Шұ ШўЩҶ ШұШ§ ШЁЩҮ ШӘШ§ШЁШ№ <code>my_any</code> ЩҲШұЩҲШҜЫҢ ШҜЩҮЫҢШҜ ЩҲ ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ ШҜШіШӘЩҲШұ Ш¬Ш§ШҜЩҲЫҢЫҢ <code>timeit</code> ЩҲ ЩҮЩ…ЪҶЩҶЫҢЩҶ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ ШўШұЪҜЩҲЩ…Ш§ЩҶ <code dir=ltr>-o</code> ШіШұШ№ШӘ Ш§Ш¬ШұШ§ЫҢ ШўЩҶ ШұШ§ Щ…ШӯШ§ШіШЁЩҮ ЩҲ ШҜШұ Щ…ШӘШәЫҢШұ <code>a</code> Ш°Ш®ЫҢШұЩҮ Ъ©ЩҶЫҢШҜ. ШіЩҫШі ШЁЩҮвҖҢШ·ШұЫҢЩӮ Щ…ШҙШ§ШЁЩҮ ШіШұШ№ШӘ Ш§Ш¬ШұШ§ЫҢ ШӘШ§ШЁШ№ <code>any</code> ШұШ§ Щ…ШӯШ§ШіШЁЩҮ Ъ©ШұШҜЩҮ ЩҲ ШҜШұ Щ…ШӘШәЫҢШұ <code>b</code> Ш°Ш®ЫҢШұЩҮ Ъ©ЩҶЫҢШҜ.
</font>
</p>

```python
x = [1,2,3]
a = %timeit -o my_any(x)
```

```python
b = %timeit -o any(x)
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШҙШ§ЫҢШ§ЩҶ Ш°Ъ©Шұ Ш§ШіШӘ Ъ©ЩҮ Щ…ЫҢвҖҢШӘЩҲШ§ЩҶЫҢШҜ ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ ШҜЩҲ Ш№Щ„Ш§Щ…ШӘ ШіЩҲШ§Щ„ (<code>??</code>) ШҜШұ ЩӮШ§ШЁЩ„ЫҢШӘ <code>introspection</code> ШӘЩҲШ¶ЫҢШӯШ§ШӘ ЩҲ ШіЩҲШұШі Ъ©ШҜ ЫҢЪ© ШӘШ§ШЁШ№ ШұШ§ ШЁШЁЫҢЩҶЫҢШҜ:
</font>
</p>

```python
??my_any
```

```python
??any
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ ШҜШіШӘЩҲШұ Ш¬Ш§ШҜЩҲЫҢЫҢ <code>save</code> Ъ©ШҜ ШӘШ§ШЁШ№ Ш®ЩҲШҜ ШұШ§ ШҜШұ ЩӮШ§Щ„ШЁ ЩҒШ§ЫҢЩ„ <code dir=ltr>.py</code> Ш°Ш®ЫҢШұЩҮ Ъ©ЩҶЫҢШҜ ШӘШ§ ШЁШӘЩҲШ§ЩҶЫҢЩ… ШҜШұ ШіЩ„ЩҲЩ„ Ш¬ЩҲШ§ШЁвҖҢШіШ§ШІШҢ ШўЩҶ ШұШ§ ШҜШұ ЩҒШ§ЫҢЩ„ <code>result.zip</code> ЩӮШұШ§Шұ ШҜЩҮЫҢЩ….
</font>
</p>


```python
from inspect import getsource

%save my_any getsource(my_any)
```


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ЩҒШұШ§Щ…ЩҲШҙ ЩҶЪ©ЩҶЫҢШҜ Ъ©ЩҮ ШЁШ§ <code>introspection</code> Щ…ЫҢвҖҢШӘЩҲШ§ЩҶЫҢШҜ Ш§Ш·Щ„Ш§Ш№Ш§ШӘ Ъ©Ш§Щ…Щ„ЫҢ ШҜШұ Щ…ЩҲШұШҜ ЩҮШұ ЪҶЫҢШІЫҢ ШЁЩҮ ШҜШіШӘ ШўЩҲШұЫҢШҜ! ЩҫШі Ш§ЪҜШұ ШҜШіШӘЩҲШұ <code>save</code> ШҜШұ ШіЩ„ЩҲЩ„ ЩӮШЁЩ„ ШЁШұШ§ЫҢвҖҢШӘШ§ЩҶ ЩҶШ§ШўШҙЩҶШ§ ШЁЩҲШҜ Ъ©Ш§ЩҒЫҢШіШӘ ШЁЩҮ Ъ©Щ…Ъ© ШҜШіШӘЩҲШұ ШІЫҢШұ Ш§Ш·Щ„Ш§Ш№Ш§ШӘ ШўЩҶ ШұШ§ Ш§ШіШӘШ®ШұШ§Ш¬ ЩҲ Щ…Ш·Ш§Щ„Ш№ЩҮ Ъ©ЩҶЫҢШҜ.

</font>
</p>


```python
%save?
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
<b>ШіЩ„ЩҲЩ„ Ш¬ЩҲШ§ШЁвҖҢШіШ§ШІ</b>
</font>
</h2>


<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    ШЁШұШ§ЫҢ ШіШ§Ш®ШӘЩҮвҖҢШҙШҜЩҶ ЩҒШ§ЫҢЩ„ <code>result.zip</code> ШіЩ„ЩҲЩ„ ШІЫҢШұ ШұШ§ Ш§Ш¬ШұШ§ Ъ©ЩҶЫҢШҜ. ШӘЩҲШ¬ЩҮ ШҜШ§ШҙШӘЩҮ ШЁШ§ШҙЫҢШҜ Ъ©ЩҮ ЩҫЫҢШҙ Ш§ШІ Ш§Ш¬ШұШ§ЫҢ ШіЩ„ЩҲЩ„ ШІЫҢШұ ШӘШәЫҢЫҢШұШ§ШӘ Ш§Ш№Щ…Ш§Щ„ ШҙШҜЩҮ ШҜШұ ЩҶШӘвҖҢШЁЩҲЪ© ШұШ§ Ш°Ш®ЫҢШұЩҮ Ъ©ШұШҜЩҮ ШЁШ§ШҙЫҢШҜ (<code>ctrl+s</code>) ШӘШ§ ШҜШұ ШөЩҲШұШӘ ЩҶЫҢШ§ШІ ШЁЩҮ ЩҫШҙШӘЫҢШЁШ§ЩҶЫҢ Ш§Щ…Ъ©Ш§ЩҶ ШЁШұШұШіЫҢ Ъ©ШҜ ШҙЩ…Ш§ ЩҲШ¬ЩҲШҜ ШҜШ§ШҙШӘЩҮ ШЁШ§ШҙШҜ.
</font>
</p>

```python
import zlib
import zipfile 
import joblib
joblib.dump(a,"a")
joblib.dump(b,"b")

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ['my_any.py', 'Ipython.ipynb', 'a', 'b']
compress(file_names)
```
