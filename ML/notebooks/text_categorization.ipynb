<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
در هنگام کار با داده‌های متنی، پیش‌پردازش داده‌ها به کمک تکنیک‌های پردازش زبان طبیعی یکی از مراحل بسیار تاثیرگذار در یادگیری مدل و عملکرد نهایی است.
در تمرین «کامنت‌کاوی» فصل دسته‌بندی با چندین تکنیک رایج پیش‌پردازش خصوصاً برای زبان فارسی آشنا شدید. در این قسمت نیز می‌توانید تابعی بنویسید که یک رشته (<code>string</code>) را گرفته، اصلاحات موردنظر شما رو روی متن اعمال کرده و در نهایت نتیجه را با فرمت یک رشته (<code>string</code>) خروجی دهد. بررسی و تحلیل کلمات موجود در مجموعه‌داده از نظر تعداد رخداد نیز می‌تواند شما را در پیش‌پردازش بهتر یاری کند.
</font>
</p>


```python
Y = train["category"].copy()
```

```python
# To-Do
from hazm import stopwords_list
stp_words = stopwords_list()
train["domain"].fillna(train["url"],inplace=True)
test["domain"].fillna(train["url"],inplace=True)

txt_clm_name = ["description","text_content","title","h1","h2"]
for name in txt_clm_name:
    train[name].fillna(" ",inplace=True)
    test[name].fillna(" ",inplace=True)

    
train["texts"] = train["description"] + ' '+ train["text_content"]+ ' '+ train["title"]+ ' '+ train["h1"]+ ' '+ train["h2"]
test["texts"] = test["description"] + ' '+ test["text_content"]+ ' '+ test["title"]+ ' '+ test["h1"]+ ' '+ test["h2"]
```

```python
train.drop(columns=['description', 'text_content', 'title', 'h1', 'h2', 'url','id'],inplace=True)
test.drop(columns=['description', 'text_content', 'title', 'h1', 'h2', 'url','id'],inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
np.random.seed(34)
ord_encoder = OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=np.random.randint(850,900))

ord_encoder.fit(train["domain"].to_numpy().reshape(-1,1))

train["domain"] = ord_encoder.transform(train["domain"].to_numpy().reshape(-1,1))
test["domain"] = ord_encoder.transform(test["domain"].to_numpy().reshape(-1,1))
```

```python
test
```

```python
train["#content"] = train.groupby("domain").count().loc[train["domain"]]["texts"].to_numpy()
test["#content"] = test.groupby("domain").count().loc[test["domain"]]["texts"].to_numpy()
```

```python
train
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
در هنگام تعبیه‌سازی به کمک کتابخانه‌ی <code>scikit-learn</code>
از آنجا که ممکن است توکن‌ساز (Tokenizer) مورد استفاده در این کتابخانه مناسب زبان فارسی نباشد، بهتر است از توکن‌سازهای مناسب این زبان استفاده کنیم. کافیست تابعی بنویسید که یک رشته (<code>string</code>) را گرفته و به کمک کتابخانه‌ی موردنظر شما (مثل <code>word_tokenize</code> در کتابخانه‌ی هضم) توکن‌های آن را جدا کند. خروجی تابع لیستی از توکن‌ها خواهد بود.
</font>
</p>


```python
from hazm import WordTokenizer
def tokenizer(text):
    w_tokenizer = WordTokenizer()
    
    return w_tokenizer.tokenize(text)
```

```python
# s = "به یک تمرین صنعتی از کاربرد یادگیری ماشین در پردازش زبان طبیعی (NLP) خوش آمدید. در این تمرین داده‌های واقعی وب فارسی که توسط پلتفرم یکتانت پالایش و جمع‌آوری شده در اختیار ما قرار گرفته است."
texts = train["texts"][0:20].apply(lambda s: tokenizer(s))
texts
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# vectorizer = CountVectorizer(input = 'content',lowercase=False ,tokenizer=lambda x: x ,analyzer='word' , stop_words=stp_words)
vectorizer = CountVectorizer(analyzer='word' , stop_words=stp_words)

# vectorizer.fit(texts)

# sample = vectorizer.transform(texts)
# sample

vectorizer.fit(train["texts"])

count_vect = vectorizer.transform(train["texts"])
count_vect_t = vectorizer.transform(test["texts"])
count_vect.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# vectorizer = CountVectorizer(input = 'content',lowercase=False ,tokenizer=lambda x: x ,analyzer='word' , stop_words=stp_words)
vectorizer = TfidfVectorizer(analyzer='word' , stop_words=stp_words)

# vectorizer.fit(texts)

# sample = vectorizer.transform(texts)
# sample

vectorizer.fit(train["texts"])

tfidf_vect = vectorizer.transform(train["texts"])
tfidf_vect_t = vectorizer.transform(test["texts"])
tfidf_vect
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=21)
pca.fit(tfidf_vect.toarray())
transformed_text = pca.transform(tfidf_vect.toarray())
transformed_text_t = pca.transform(tfidf_vect_t.toarray())
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
توجه داشته باشید انتخاب این‌که از کدام ویژگی‌ها (متن‌ها) به‌عنوان ورودی الگوریتم استفاده کنید بر عهده‌ی خودتان است. می‌توانید تنها از یک ستون مثل <code>title</code> استفاده کنید یا می‌توانید به‌نحوی متن یا حتی بردار ویژگی هر ستون را با همدیگر ترکیب کنید. همچنین فراموش نکنید که ستون متغیر هدف یعنی <code>category</code> نیاز به کدگذاری دارد.
</font>
</p>


```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
target = encoder.fit_transform(train["category"])
Y = encoder.transform(Y)
train.drop(columns=["category"],inplace=True)
```

```python
pd.DataFrame(target).value_counts()
```

```python
from sklearn.model_selection import train_test_split
df, validatin_text, target, validation_target = train_test_split(transformed_text,target , test_size=0.1)
```

مدل‌سازی

</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
اکنون وقت آن رسیده که الگوریتم یادگیری ماشین موردنظر خود را بر روی داده‌های آموزشی اجرا کنید. در انتخاب الگوریتم کاملاً آزاد هستید. برای این بخش می‌توانید ابتدا هر ورودی متن را به کمک تکنیک‌های معرفی‌شده در درسنامه‌های این فصل (مثل Bag-of-Word یا Tf-idf) به بردارهای ویژگی عددی تبدیل کنید. سپس این بردارها را همراه با لیست برچسب‌های صحیح به الگوریتم بدهید تا مدل آموخته شود. علاوه بر این می‌توانید تمام این مراحل را در یک خط لوله‌ (<code>Pipeline</code>) نیز تعریف کنید.
</font>
</p>

```python
# To-Do
from sklearn.svm import SVC
model = SVC(C = 35,class_weight='balanced')
model.fit(df,target)
model.score(df,target)
```

```python
from sklearn.metrics import f1_score
score = f1_score(validation_target,model.predict(validatin_text),average='weighted')
score
```
